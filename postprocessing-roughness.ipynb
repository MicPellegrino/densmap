```python
import scipy.optimize as sc_opt
import scipy.special as sc_spc
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rational(x, p, q) :
    return np.polyval(p, x) / np.polyval(q + [1.0], x)
def rational_4_2(x, p0, p1, p2, p3, q1) :
    return rational(x, [p0, p1, p2, p3], [q1,])
def ratioder_4_2(x, p0, p1, p2, p3, q1) :
    return rational(x, [2*p0*q1, 3*p0+p1*q1, 2*p1, p2-p3*q1], [q1*q1, 2*q1,])
def ratiodr2_4_2(x, p0, p1, p2, p3, q1) :
    return rational(x, [2*p0*q1*q1, 6*p0*q1, 6*p0, 2*(p1-p2*q1+p3*q1*q1)], [q1**3, 3*q1*q1, 3*q1])
```

```python
cos = lambda t : np.cos( np.deg2rad(t) )
sin = lambda t : np.sin( np.deg2rad(t) )
tan = lambda t : np.tan( np.deg2rad(t) )
cot = lambda t : 1.0/np.tan( np.deg2rad(t) )
tan_m1 = lambda t : np.rad2deg(np.arctan(t))
cos_m1 = lambda t : np.rad2deg(np.arccos(t))
sin_m1 = lambda t : np.rad2deg(np.arcsin(t))
```

```python
sinh = lambda x, a, b, c : a*np.sinh(b-c*x)
```

```python
gamma = 57.8
TOCP = 1e3
muf_ref = 5.66
PS2NS=1e-3
```

```python
p0_ls=[1,1,1,1,1]
mv1=1000
mv2=10000
lvel=10
lacc=1
iavg_theta = 650
```

```python
PLOT_RES = False
```

```python
n_tag_vec = ['06','08','10','12','14','16']
a_tag_vec = ['02','04','06','08','10']
rep_folders = ['Rep0','Rep1','Rep2','Rep3','Rep4']
folders = [r[0]+r[-1] for r in rep_folders ]
print(folders)
```

```python
mu_f_vec = np.zeros( (len(n_tag_vec),len(a_tag_vec)) )
theta_fin_vec = np.zeros( (len(n_tag_vec),len(a_tag_vec)) )
i = 0

for n in n_tag_vec:
    j = 0
    
    for a in a_tag_vec:
        
        tag = 'n'+n+'a'+a
        folders_root = 'AmorphousSpreading/'+tag+'/'
        print('---------- MKT '+tag+' --------------------')
        
        t = np.loadtxt(folders_root+folders[0]+'/time.txt')*PS2NS
        nt = len(t)
        
        def rd_m(p) :
            r = ratioder_4_2(t,*p)
            r[r>0] = 0
            return r 
        def r2_p(p) :
            r = ratiodr2_4_2(t,*p)
            r[r<0] = 0
            return r
        
        fnpz1 = folders_root+'x_avg_'+tag+'.npz'
        fnpz2 = folders_root+'theta_g_'+tag+'.npz'
        
        X = np.load(fnpz1)
        t = X['x']
        xcl_avg = X['y']
        X = np.load(fnpz2)
        theta_avg = X['y']
        
        theta_fin = np.mean(theta_avg[iavg_theta:])
        theta_fin_vec[i,j] = theta_fin
        print('theta_fin = '+str(theta_fin)+' deg')
                    
        def fres(p):
            return np.concatenate((xcl_avg-rational_4_2(t,*p),lvel*rd_m(p),lacc*r2_p(p)),axis=None)

        ls_results = sc_opt.least_squares(fres,x0=p0_ls,max_nfev=mv1)
        popt1 = ls_results.x
        x_fit = rational_4_2(t,*popt1)
        v_fit = ratioder_4_2(t,*popt1)
        
        w = np.abs(v_fit)
        w /= np.sum(w)
        popt2, pcov2 = sc_opt.curve_fit(sinh,cos(theta_avg),v_fit,maxfev=mv2,sigma=np.sqrt(w))
        
        # popt2, pcov2 = sc_opt.curve_fit(sinh,cos(theta_avg),v_fit,maxfev=mv2)
        
        muf = gamma/popt2[0]*popt2[2]
        mu_f_vec[i,j] = muf
        print('muf_eff = '+str(muf)+' cP')
        print('------------------------------------------')
           
        if PLOT_RES :
            v_mkt = sinh(cos(theta_avg),*popt2)
            plt.plot(cos(theta_avg),v_fit,'k.',linewidth=3)
            plt.plot(cos(theta_avg),v_mkt,'r-',linewidth=2.5)
            plt.xlabel(r'$\theta$ [deg]')
            plt.ylabel(r'$x_{cl}$ [nm]')
            plt.show()
    
        j += 1
    i += 1
```
